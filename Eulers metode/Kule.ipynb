```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'

#Fysiske st√∏rrelser
œÄ = np.pi
g = 9.81              #Tyngdeakselerasjon i m/s/s
Œ± = œÄ/6.               #Vinkel til skr√•plan i radianer
a = g*np.sin(Œ±)        #Akselerasjonen som virker p√• bilen. (endringen i fart)

v_0 = 0                #Startfart i m/s
s_0 = 0                #Startstrekning m

#Tidsintervaller
T  = 3.05               #Tid i s
dt = 0.05               #Steglengde
N  = round(T/dt)        #Antall steg

t = np.linspace(0, T, N) #Tidsarray
v = np.zeros(N)          #Fartsarray med nuller (skal fylles ved hjelp av eulers metode)
s = np.zeros(N)          #Strekningsarray med nuller

v[0] = v_0
s[0] = s_0
fram = True

#Eulers metode
for i in range(N - 1):
    
    if Œ± < œÄ and s[i] > 1:
        Œ± = 0
        a = g*np.sin(Œ±)
        
    if Œ± == 0 and s[i] > 5:
        Œ± = œÄ+œÄ/3
        a = g*np.sin(Œ±)
        
    v[i + 1] = v[i] + a*dt     
    s[i + 1] = s[i] + v[i]*dt
    

plt.rcParams['figure.figsize'] = (12, 2.5)
plt.xlabel('$s$') # Merker x-aksen
plt.ylabel('$m/s$') # Merker y-aksen
plt.title('Fart') #Tittel
plt.plot(t, v)
plt.axhline(0, color='gray', lw=0.5)
plt.show()

plt.xlabel('$s$') # Merker x-aksen
plt.ylabel('$m$') # Merker y-aksen
plt.title('Strekning') #Tittel
plt.plot(t, s, 'r')
plt.axhline(0, color='gray', lw=0.5)
plt.show()
```
