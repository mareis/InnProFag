```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'

# ----------------    Fysiske st√∏relse  --------------------

g = 9.81                   # Gravitasjon [m/s^2]
m = 0.3                    # Massen til kula [kg]
L = 5                      # Lengden til snora
œÅ = 1.2                    # Tettheten til lufta [kg/m^3]
C_d = 0.28                 # Dragkoofesienten 
r = 2
A = 3.14*r**2            # Tversnitt til kula [m^2]
B = œÅ*C_d*A/2              # Samler konstantene i B

# ----------------    Tidsintervaller  --------------------

T  = 1             # Tid [s]
Œît = 0.0005                # Tidssteg [s]
N  = round(T/Œît)           # Antall tidssteg
x  = np.zeros(N)           # Posisjonsarray x-retning [m]
y  = np.zeros(N)           # Posisjonsarray y-retning [m]
œâ  = np.zeros(N)           # Fartsarray x-retning [m/s]
Œ∏  = np.zeros(N)           # Fartsarray y-retning [m/s]
t  = np.linspace(0, T, N)  # Tidsarray [s]


# ----------------  Startverdier  --------------------

œâ[0] = 0
Œ∏[0] = np.pi/5 
x[0] = 0
y[0] = 0

# ----------------  Eulers metode  --------------------

for i in range(N - 1):
    if i < N/6:
        Œ∏[i] = np.pi/5
        g=np.abs(g)
    elif i < N/4:
        Œ∏[i] = 0.000001
    else:
        Œ∏[i] = np.pi/7
        g=-g
    
    œâ[i + 1] = œâ[i] + 5*g/np.sin(Œ∏[i])/(7*r)
    
    x[i+1] = r*(œâ[i+1]*Œît - np.sin(œâ[i+1]*Œît))
    y[i+1] = r*(1 - np.cos(œâ[i]*Œît))
    

# -------------------  Plot  -------------------------
    
plt.rcParams['figure.figsize'] = (8, 4.5)
plt.style.use('ggplot')
plt.xlabel('tid $[s]$')     
plt.ylabel('$\\theta$ [radianer]')     
plt.title('Vinkel')          
plt.plot(t, np.rad2deg(Œ∏))
#plt.savefig('pendelbevegelse.png')
plt.show()

plt.plot(x, y)
plt.show()

```
